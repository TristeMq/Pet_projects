```python
#приведем к нужному  виду массив с тестовыми данными
df_test['lithology'] = df_test["lithology"].astype("category")
encoder = OneHotEncoder(handle_unknown='ignore')
encoder_df_test = pd.DataFrame(encoder.fit_transform(df_test[['lithology']]). toarray ())
df_test = df_test.join (encoder_df_test)
encoder_df.head()
df2 = df.drop("lithology",axis = "columns")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader, ConcatDataset
from torch.optim import SGD
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.decomposition import PCA
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

```python

class DataSet:
  def __init__(self, dataset):
    #dataset.columns = dataset.columns.astype(str)
    #pca = PCA(n_components=20)
    #c = pca.fit_transform((dataset.drop(dataset.loc[:,'bv_0':'pc_100'],axis = 1).values))
    #self.x_data = c
    #PCA  ухудшил модель, поэтому лучше его не использовать
   
    
    self.x_data = preprocessing.normalize((dataset.drop(dataset.loc[:,'bv_0':'pc_100'],axis = 1).values))
    self.y_data = preprocessing.normalize((dataset.loc[:,'bv_0':'pc_100']).values)
    self.len = len(self.x_data)
  def __getitem__(self, index):
        return self.x_data[index], self.y_data[index]
  def __len__(self):
        return self.len
x = DataSet(df1)
y = DataSet(df2)
train_loader = DataLoader(x, shuffle=True, batch_size=32)
test_loader = DataLoader(y, shuffle=False, batch_size=32)
```

```python
class MultiOutputRegression(nn.Module):
    def __init__(self):
        super(MultiOutputRegression, self).__init__()
        self.model = nn.Sequential(
        nn.Linear(24, 15),
        nn.ReLU(),
        nn.Linear(15, 10),
        nn.ReLU(),
        nn.Linear(10,202))
        
        
    def forward(self, x):
        return self.model(x)
model = MultiOutputRegression()
criterion = nn.MSELoss()
#optimizer = torch.optim.SGD(model.parameters(),lr=0.000000000000000000001, momentum=0.9)# попробуй другие оптимайзеры
optimizer = torch.optim.Adam(model.parameters())
print(model)
```

```python
PATH = "init.pth"
torch.save(model.state_dict(), PATH)
```

```python
def plot_losses(train_losses, test_losses):
    fig, axs = plt.subplots(figsize=(13, 4))
    axs.plot(range(1, len(train_losses) + 1), train_losses, label='train')
    axs.plot(range(1, len(test_losses) + 1), test_losses, label='test')
    axs.set_ylabel('loss')
    axs.set_xlabel('epoch')
    axs.legend()

    plt.show()
```

```python

train_losses = []
test_losses = []
for epoch in range(20):
  train_loss = 0
  for x_batch, y_batch in train_loader:
          x_batch = x_batch.to(device)
          y_batch = y_batch.to(device)

          model.train()
          optimizer.zero_grad()
          y_pred = model(x_batch.to(torch.float32))
          loss = torch.sqrt(criterion(y_batch.to(torch.float32),y_pred))#RMSE Loss
          loss.backward()
          optimizer.step()
          train_loss += loss.item() 
  epoch_loss = train_loss/len(test_loader.dataset)
  train_losses.append(epoch_loss)
  test_loss = 0
  for x_batch, y_batch in test_loader:

          x_batch = x_batch.to(device)
          y_batch = y_batch.to(device)

          with torch.no_grad():
            y_pred = model(x_batch.to(torch.float32))
            loss = torch.sqrt(criterion(y_batch.to(torch.float32),y_pred))#RMSE Loss
          test_loss += loss.item() 
  n_test = len(test_loader.dataset)
  test_loss = test_loss/n_test
  test_losses.append(test_loss)
 

         


 

plot_losses(train_losses,test_losses)
```

Оценим модель при помощи k-fold Cross-validation 

```python
from tqdm.notebook import tqdm, trange
def train(dataloader, model):
    size = len(dataloader.dataset) 
    num_batches = len(dataloader) 

    model.train() 
    epoch_loss  = 0

    for batch_i, (x, y) in enumerate(tqdm(dataloader, leave=False)):
        x, y = x.to(device), y.to(device) 
        
        pred = model(x.to(torch.float32))
        loss = torch.sqrt(criterion(pred, y.to(torch.float32)))#RMSE Loss

        
        optimizer.zero_grad() 
        loss.backward() 
        optimizer.step() 

       
        epoch_loss += loss.item()

 
    return epoch_loss/num_batches
    

def test(dataloader, model):
    size = len(dataloader.dataset)
    num_batches = len(dataloader) 

    model.eval() 
    epoch_loss = 0

 
    with torch.no_grad():
        for batch_i, (x, y) in enumerate(tqdm(dataloader, leave=False)):
            x, y = x.to(device), y.to(device)

            pred = model(x.to(torch.float32))
            loss = torch.sqrt(criterion(pred, y.to(torch.float32)))#RMSE Loss

            epoch_loss += loss.item()
    return epoch_loss/num_batches
```

```python
kfold = KFold(n_splits = 10)
from sklearn.model_selection import KFold

fold_losses = []
fold_accs = []

for fold_i, (train_ids, val_ids) in enumerate(kfold.split(x)):
    print(f'train size: {len(train_ids)}, val size: {len(val_ids)}')

  
    model.load_state_dict(torch.load(PATH))

   
    train_sampler = torch.utils.data.SubsetRandomSampler(train_ids)
    val_sampler = torch.utils.data.SubsetRandomSampler(val_ids)
   
    train_loader = DataLoader(x, batch_size=25, sampler=train_sampler)
    val_loader = DataLoader(x, batch_size=25, sampler=val_sampler)
    
    optimizer = torch.optim.Adam(params=model.parameters())
 
    for epoch in tqdm(range(5), leave=False):
        train_loss = train(train_loader, model)
        val_loss = test(val_loader, model)
    
  
    test_loss = test(test_loader, model)
    print(f'Fold {fold_i}, test loss: {test_loss:.3f}')

    fold_losses.append(test_loss)
```

```python
print(f"Loss: mean {np.mean(fold_losses):.3f}, std: {np.std(fold_losses):.3f}")
```

#Как можно улучшить модель:
1.Добавить новых признаков (вручную или через polynomial features), однако в таком случае может возникнуть проблема переобучения ( в этом случае можно использовать регуляризацию, в случае этой модели лучше всего 
drop out)
2. При кросс-валидации лучше всего проводить нормализацию в каждой части  отдельно, чтобы избежать влияния всего датасета на отдельную выборку
3.В случае обучения модели на большем количестве  данных  можно использовать batch-normalization
4. Также можно в дальнейшем эксперементировать со структурой нейросети, однако при усложнении модели вероятно увеличение времени обучения
5. Конкретно в данной ситуации метод главных компонент PCA привел к ухудшению качества модели, однако можно попробовать другие методы feature selection(VarianceThreshold, Forward Selection etc)
6. Отдельно можно попробовать провести сравнительный анализ между этой моделью и различными бустингами и понять примерное соотношение качества модели к вычислительным сложностям (опять же актуальнее в случае обучения на большом датасете)
